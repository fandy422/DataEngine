```python
import pandas as pd
from pandas import DataFrame as df
csv_data = pd.read_csv('car_complain.csv') #读取数据
```

```python
csv_data.head(10) #显示数据头10行
```

```python
csv_data=csv_data.drop('problem',axis=1).join(csv_data['problem'].str.get_dummies(","))#拆分Problem
csv_data 
```

```python
csv_data['brand']=csv_data['brand'].apply(lambda x: x.replace("-","") if "-" in x else x)  #替换-字符
csv_data_brand=csv_data.groupby(by='brand')['id'].agg(['count'])  #品牌投诉总数
csv_data_brand
```

```python
csv_data_model=csv_data.groupby(by='car_model')['id'].agg(['count'])  #车型投诉总数
csv_data_model
```

```python
tags=csv_data.columns[7:]  #问题标签
tags
```

```python
csv_data_result=csv_data.groupby(by='brand')[tags].agg(['sum'])  #品牌对应各投诉问题详情
csv_data_result
```

```python
csv_data_result=csv_data_brand.merge(csv_data_result,left_index=True,right_index=True,how='left')  #合并后的品牌投诉
csv_data_result
```

```python
csv_data_result.reset_index(inplace=True)  #生成index
csv_data_result
```
